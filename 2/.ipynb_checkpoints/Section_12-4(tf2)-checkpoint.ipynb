```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 최소값, 최대값이 각각 0, 1이 되고, 사이의 값은 0~1 사이의 값으로 set.
# 값이 들쑥날쑥하기 때문에 적용해줘야 한다.
# Normalize
def MinMaxScaler(data):
    numerator = data - np.min(data, 0)
    denominator = np.max(data, 0) - np.min(data, 0)
    # noise term prevents the zero division
    return numerator / (denominator + 1e-7)
```

```python
seq_length = 7 #7일간 data
data_dim = 5
output_dim = 1
learning_rate = 0.01
iterations = 500
```

```python
xy = np.loadtxt('data-02-stock_daily.csv', delimiter=',')
# 시간순으로 만들기 위해서 값을 뒤집는다.
xy = xy[::-1] #reverse order(chronically ordered)
```

```python
train_size = int(len(xy) * 0.7) #70%만 사용
train_set = xy[0:train_size]
test_set = xy[train_size-seq_length:] 
#과거의 순서를 사용하기 위해 train_size-seq_length를 함
```

```python
train_set = MinMaxScaler(train_set)
test_set = MinMaxScaler(test_set)
```

```python
def build_dataset(time_series, seq_length):
    dataX = []
    dataY = []
    for i in range(0, len(time_series)-seq_length):
        x = time_series[i:i+seq_length, :]
        y = time_series[i+seq_length, [-1]]
        #print(x, '->', y)
        dataX.append(x)
        dataY.append(y)
    return np.array(dataX), np.array(dataY)
```

```python
trainX, trainY = build_dataset(train_set, seq_length)
testX, testY = build_dataset(test_set, seq_length)
print(trainX.shape, trainY.shape)
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.LSTM(
    units=1, input_shape=(seq_length, data_dim),))
tf.model.add(tf.keras.layers.Dense(units=output_dim,activation='tanh'))
tf.model.summary()
```

```python
tf.model.compile(loss='mean_squared_error',
                optimizer=tf.keras.optimizers.Adam(lr=learning_rate))
```

```python
tf.model.fit(trainX, trainY, epochs=iterations)
```

```python
evaluation = tf.model.evaluate(trainX, trainY)
print('\t##Loss:', round(evaluation,7))
```

```python
plt.plot(trainY)
plt.plot(tf.model.predict(trainX))
plt.show()
```

```python
test_predict = tf.model.predict(testX)
```

```python
plt.plot(testY, color='green')
plt.plot(test_predict, color='red')
plt.xlabel('Time Priod')
plt.ylabel('Stock Price')
plt.show()
```
